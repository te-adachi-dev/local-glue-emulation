```python
import sys
import os
from pyspark.context import SparkContext
from pyspark.sql import SparkSession
from awsglue.context import GlueContext
from awsglue.job import Job
from awsglue.utils import getResolvedOptions

# Glueジョブ初期化
sc = SparkContext()
glueContext = GlueContext(sc)

# Sparkセッションの拡張設定
spark = SparkSession.builder \
    .appName("JupyterGlueJob") \
    .config("hive.metastore.uris", "thrift://hive-metastore:9083") \
    .config("spark.sql.warehouse.dir", "/home/glue_user/workspace/spark-warehouse") \
    .enableHiveSupport() \
    .getOrCreate()

# データベースとテーブルの指定
database = "test_db_20250407"
table = "test_table"

# Hiveテーブルからデータ読み込み
df = spark.sql(f"SELECT * FROM {database}.{table}")
df.show()

# 処理例：年齢が30歳以上のレコードをフィルタリング
filtered_df = df.filter(df.age >= 30)
filtered_df.show()

# 結果をHiveテーブルとして保存
output_table = f"{table}_filtered"
filtered_df.write.mode("overwrite").saveAsTable(f"{database}.{output_table}")

print(f"データを処理して {database}.{output_table} に保存しました")
```
